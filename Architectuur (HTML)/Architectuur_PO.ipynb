Uitleg Eigen website:
        Maak een website met verschillende webpagina’s waarin je laat zien wat je geleerd hebt. Maak de website door zelf de code in HTML en CSS te schrijven in.  Alles wat behandeld wordt tijdens de les en in OneNote staat is onderdeel van de stof. De opgaven die in OneNote staan kun je zelf gebruiken als oefening, die hoef je niet in te leveren. 
        <hr>

In deze opdracht ga je jouw website verder uitbouwen.




Als het goed is heb je <a href="#PO5A">PO TAAK 5A: WEBSITE BOUWEN (DEEL 1)</a> af. Zo niet, maak die dan eerst, want die heb je nodig.


a) Maak in VSC een mapje aan voor de code van je eigen website, bv 'POeigenWebsite'.

b) Maak binnen je website 3 nieuwe lege HTML-pagina's aan. Geef ze zinvolle namen, gebruik hiervoor de namen dat je bij <a href="#PO2"><i>PO TAAK 2. WEBSITE ONTWERPEN</i></a> had bedacht.


c) Voeg een menu toe aan je hoofdpagina, net als in paragraaf <a href="Architectuur_H4.ipynb#H4.5">4.5 Menubalk</a> is uitgelegd.

d) Test je menu door op elke link te drukken.

e) Zodra je een deel werkend hebt, maak je gebruik van versiebeheer, zoals in paragraaf <a href="Architectuur_H1.ipynb#H1.9">Versiebeheer</a> is uitgelegd. Geef je commit een zinvolle label, bijvoorbeeld '<i>frame af & getest</i>')


f) Kopieer de code van je hoofdpagina naar de andere pagina's. Pas voor elke pagina de tekst en de menu aan. Test alle links. Maak gebruik van versiebeheer.

Programmeertools:
        • Programmeromgeving: Maak gebruik van de beschikbare repl.it: https://replit.com/team/ProgrammereninP/POeigenWebsite
        • Versiebeheer: Maak structureel gebruik van versiebeheer met zinvolle labels (bv. Heb je de frame van je ontwerp juist in code omgezet, doe dan een commit met de tekst "frame af & getest")
<hr>
Vereisten voor de website:
        • De website bestaat uit minstens 4 pagina's.
        • De website heeft  één hoofdpagina: index.html
        • De website heeft  één stylesheet: style.css.
        • Elke pagina heeft een menu waarmee je kunt doorklikken naar alle andere pagina's. 
        • Je moet minstens één geordende lijst (ol) en één ongeordende lijst (ul) hebben (bijvoorbeeld een profielpagina met je hobby’s, of je favoriete vakantiebestemmingen).
        • Een fotopagina met minimaal 4 foto's in een tabel waarbij elke foto klikbaar is en dus verwijst naar een andere website of pagina (bijvoorbeeld: als je foto is gemaakt in Spanje klikt, gaat deze naar de wikipedia van Spanje). Minimaal 1 externe link (naar een andere website). Denk ook aan wat er getoond wordt als je foto niet kan laden. Probeer aan te sluiten bij het onderwerp van je website.
        • Je HTML en CSS code is netjes gestructureerd en voorzien van commentaar.
        • Je project bevat geen code dat niet gebruikt wordt (denk er aan om je CSS op te ruimen, gooi ook alle ongebruikte bestanden weg).

Uitbreidingen
Hoe meer je laat zien wat je geleerd hebt, hoe hoger je cijfer. Haal dus al je kennis uit de kast! 
Ga naast de aangeboden stof ook zelf op zoek gaan naar verdieping. Op internet heel veel informatie te vinden. Probeer nieuwe dingen uit! Maak fouten en leer daarvan, zo leer je beter hoe het echt in elkaar zit (en als je daarover reflecteerd krijg je nog een hoger cijfer ook!)
Let er wel op dat je niet zomaar dingen moet knippen en plakken zonder dat je begrijpt wat het doet. Als je het niet helemaal begrijpt, moet je het niet gebruiken! Leen je iets, maak dan gebruik van bronvermelding en geef in commentaar aan wat het doet en hoe het werkt. Alle code die in je website voorkomt moet je kunnen uitleggen.
