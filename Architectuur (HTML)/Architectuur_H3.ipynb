# <span style="color:brown">3.1: Ontwerp vuistregels</span>

<table><tr><td>

Bij het ontwikkelen van een website staat de gebruiker bovenaan. Is de website niet gebruikersvriendelijk, dan klikt de gebruiker door naar een ander website en komt niet meer terug.

Jakob Nielsen ontwikkelde de onderstaande 10 zogenaamde heuristieken (vuistregels) om de gebruikersvriendelijkheid van een website te waarborgen:

## <span style="color:brown">Heuristieken (van Jakob Nielsen)</span>

<table>
<tr bgcolor="#F5CBA7"><th width="50%">Heusistieken</th><th>Voorbeeld</th></tr>

<tr style="vertical-align:top" "><td>
<span style="color:brown"><b>1: Laat zien wat er gebeurt</b></span><br>
Laat je bezoeker altijd weten waarom hij bijvoorbeeld moet wachten en hoe lang het gaat duren. Denk aan het aantal stappen dat de bezoeker nog moet doorlopen. Laat bij het aankoopproces een stappenplan zien, hoe ver de consument is en hoeveel stappen hij nog moet doen. Het is als wachten voor een stoplicht. Wanneer er wordt afgeteld, is men geduldiger dan wanneer dit niet het geval is.

4: Wees consistent</b></span><br>
Gebruik elementen en acties waar je bezoeker aan gewend is. Dat kunnen bepaalde woorden zijn, maar ook waar dat je bepaalde functies en layout van een pagina plaatst. Conventies geven een vertrouwd gevoel. Bepaalde onderdelen worden altijd verwacht door bezoekers. Bv. een logo, een zoekfunctie of het taalgebruik waarmee je bezoeker wordt aangesproken. Het maakt dat bezoekers zich goed voelen met de huidige standaarden. De belangrijkste plaats voor een website is rechtsboven, want daar kijkt de bezoeker het eerst. # <span style="color:brown">3.2: Pagina opbouw met blokken</span>

<table><tr><td>

Als je een website wilt maken, begin je niet direct met het schrijven van code in HTML en CSS. Je begint met een goed ontwerp. In het ontwerpproces denk je niet alleen na over de inhoud, afbeeldingen en kleuren die je gaat gebruiken. Een van de belangrijkste onderdelen van het ontwerpen is het nadenken over de structuur van elke pagina. Het is voor de gebruiker prettig als de structuur van de verschillende pagina's op de website hetzelfde blijft. Om een goede structuur te krijgen, moet je leren '<i>denken in blokken</i>'.

Websites zijn opgebouwd uit blokken. Er is een blok voor het logo, voor de navigatie, voor het belangrijkste onderwerp, voor tekst, enzovoorts. Om je daarvan een goed beeld te geven, zie je hier een paar voorbeelden van websites waar je dit goed kunt zien. Let eens goed op de manier waarop bij deze websites alle onderdelen geplaatst zijn. # <span style="color:brown">3.3: Ontwerp maken</span>

<table><tr><td>

Voordat je begint met het maken van een website bepaal je hoe je pagina's logisch worden ingedeeld en eenduidig worden opgemaakt. Heeft je website meerdere pagina's, dan houd je dezelfde indeling en opmaak aan op alle pagina's, dat zorgt voor eenduidigheid. 

Een ontwerp bevat een beschrijving van de:

a) Pagina <b>layout</b> van de hoofdpagina: een schets van de elementen met hun breedte en hoogte, en waar bijvoorbeeld een bedrijfsnaam of logo komt te staan.

b) <b>Opmaakstijlen</b> voor alle pagina's: 
- Gebruikte kleuren: denk aan achtergrond, knopjes, koppen;
- Tekstopmaak: denk aan lettertype, letterkleuren.
