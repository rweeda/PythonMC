# <span style="color:brown">3.1: Ontwerp vuistregels</span>

Bij het ontwikkelen van een website staat de gebruiker bovenaan. Is de website niet gebruikersvriendelijk, dan klikt de gebruiker door naar een ander website en komt niet meer terug.

Jakob Nielsen ontwikkelde de onderstaande 10 zogenaamde heuristieken (vuistregels) om de gebruikersvriendelijkheid van een website te waarborgen:

## <span style="color:brown">Heuristieken (van Jakob Nielsen)</span>

<table>
<tr><th>Heusistieken</th><th>Voorbeeld</th></tr>

<tr style="vertical-align:top"><td>
<span style="color:brown"><b>1: Laat zien wat er gebeurt</b></span><br>
Laat je bezoeker altijd weten waarom hij bijvoorbeeld moet wachten en hoe lang het gaat duren. Denk aan het aantal stappen dat de bezoeker nog moet doorlopen. Laat bij het aankoopproces een stappenplan zien, hoe ver de consument is en hoeveel stappen hij nog moet doen. Het is als wachten voor een stoplicht. Wanneer er wordt afgeteld, is men geduldiger dan wanneer dit niet het geval is.
</td><td>
Een timer laat zien hoe lang de gebruiker nog moet wachten:
<img src="img/3.1_heuristiek1.png" width="200" alt="Heustiek 1: Laat zien wat er gebeurt">
</td></tr>

<tr style="vertical-align:top"><td>
<span style="color:brown"><b>2: Het systeem en de echte wereld komen overeen</b></span><br>
Spreek dezelfde taal en beeldtaal als je bezoeker. Gebruik termen die hij of zij ook gebruikt en de afbeeldingen en icoontjes waaraan men is gewend. Dit is belangrijk voor de beleving van de bezoeker.
</td><td>
Gebruik termen die je bezoekers begrijpen, jongerentaal of vaktaal bij hoogopgeleiden:<br>

<img src="img/3.1_heuristiek2a.png" width="600" alt="Heustiek 2: Het systeem en de echte wereld komen overeen">
<br>

De knopjes om de gaspitten te bedienen staan in dezelfde volgorde als de gastpitten zelf:<br>
<img src="img/3.1_heuristiek2b.png" width="200" alt="Heustiek 2: Het systeem en de echte wereld komen overeen">
</td></tr>
<tr style="vertical-align:top"><td>
<span style="color:brown"><b>
3: Gebruiker heeft controle en vrijheid</b></span><br>
Zorg dat je bezoeker beslissingen ongedaan kan maken. De website zou de bezoeker altijd een gemakkelijke uitweg moeten bieden. Deze controle geeft je bezoeker een beter vertrouwen en meer zekerheid.
</td><td>
Een pijltje om terug te keren naar de vorige pagina of om iets ongedaan te maken:<br>

<img src="img/3.1_heuristiek3.png" width="300" alt="Heustiek 3: Gebruiker heeft controle en vrijheid">
</td></tr>

<tr style="vertical-align:top"><td>
<span style="color:brown"><b>
4: Wees consistent</b></span><br>
Gebruik elementen en acties waar je bezoeker aan gewend is. Dat kunnen bepaalde woorden zijn, maar ook waar dat je bepaalde functies en layout van een pagina plaatst. Conventies geven een vertrouwd gevoel. Bepaalde onderdelen worden altijd verwacht door bezoekers. Bv. een logo, een zoekfunctie of het taalgebruik waarmee je bezoeker wordt aangesproken. Het maakt dat bezoekers zich goed voelen met de huidige standaarden. De belangrijkste plaats voor een website is rechtsboven, want daar kijkt de bezoeker het eerst. Het logo staat meestal linksboven op de website en het menu staat meestal ook boven aan.
</td><td>
Een winkelmand is te herkennen aan een standaard icoon. Het winkelmadje is vaak rechts bovenaan een pagina te vinden:<br>

<img src="img/3.1_heuristiek4.png" width="100" alt="Heustiek 3: Gebruiker heeft controle en vrijheid">
</td></tr>

<tr style="vertical-align:top"><td>
<span style="color:brown"><b>
5: Wees kritisch en bied hulp</b></span><br>
Fouten zijn belangrijk om te voorkomen. Het aangeven van een foutmelding is altijd snel gedaan, maar als je bezoeker niet weet wat hij fout heeft gedaan zal hij de ervaring as vervelend ervaren en je pagina verlaten. Je probeert de gebruiker te behoeden van eventuele fouten en te helpen met het aangeven. Dit geeft een gevoel van zekerheid bij de gebruiker.

</td><td>
Controle op geldige invoer, en geven van een aanwijzing om deze te corrigeren:<br>


<img src="img/3.1_heuristiek5.png" width="500" alt="Heustiek 5: Wees kritisch en bied hulp">
</td></tr>

<tr style="vertical-align:top"><td>
<span style="color:brown"><b>
6: Zorg dat de bezoeker niks hoeft te onthouden</b></span><br>
Zorg dat je website de dingen onthoudt die belangrijk zijn, of vermeld de informatie gewoon nog een keer daar waar nodig is.
</td><td>
Als je in Word een bestand wil openen, dan worden recent geopende documenten getoond waar je uit kan kiezen:
<img src="img/3.1_heuristiek6.png" width="300" alt="Heustiek 6: Zorg dat de bezoeker niks hoeft te onthouden">
</td></tr>


<tr style="vertical-align:top"><td>
<span style="color:brown"><b>7: Flexibiliteit en efficiency</b></span><br>
Je website moet afgestemd te zijn op je bezoeker. Nieuwe bezoekers lezen eerder de uitleg dan een ervaren bezoeker. Ervaren bezoekers willen zelf meer controle kunnen uitvoeren.
</td><td>
De functie ‘geavanceerde instellingen’ op je computer is voor de gevorderde of specialist.
<img src="img/3.1_heuristiek7.png" width="350" alt="Heustiek 7: Flexibiliteit en efficiency.">
</td></tr>

<tr style="vertical-align:top"><td>
<span style="color:brown"><b>
8: KIS (Keep-It-Simple)</b></span><br>
Google is het perfecte voorbeeld van een simpele en eenvoudige site. Zorg ervoor dat je bezoeker zijn taak op je website kan uitvoeren. Door eenvoud weet de gebruiker hou hij te werk moet gaan. Dit geldt zowel voor functionaliteiten als teksten als de vormgeving.

</td><td>
Schrap overbodige tekst en vormgeving:
<img src="img/3.1_heuristiek8.png" width="300" alt="Heustiek 8: KIS (Keep-It-Simple).">
</td></tr>

<tr style="vertical-align:top"><td>
<span style="color:brown"><b>
9: Maak foutmeldingen minder eng</b></span><br>
Maak van foutmeldingen netjes een melding. Vertel wat er aan de hand is. Leg de schuld niet bij je bezoeker en zorg voor een makkelijke vluchtroute. Zie het ook als een mogelijkheid om er wat leuks van de maken.
</td><td>
Bij Mailchimp krijg je een leuke foutmelding met een grappig plaatje als je je wachtwoord of gebruikersnaam verkeerd intypt:
<img src="img/3.1_heuristiek9.png" width="300" alt="Heustiek 9: Maak foutmeldingen minder eng">
</td></tr>

<tr style="vertical-align:top"><td>
<span style="color:brown"><b>
10: Bied een helpende hand</b></span><br>
Eigenlijk moet je website al helemaal duidelijk zijn zonder hulp. Bied dus alleen extra hulp als het nodig is. Het is eigenlijk extra service.
</td><td>
Een chat-functie rechtsonderin voor het stellen van vragen:
<img src="img/3.1_heuristiek10.png" width="200" alt="Heustiek 10: Bied een helpende hand">
</td></tr>
