## <span style="color:brown">Data opslag</span>
Data wordt meestal op een gestructureerde manier opgelsagen. Door het zo efficiënt mogelijk op te slaan is de data makkelijk te vinden en te bewerken.

Data wordt meestal opgeslagen in een database. Een database is eigenlijk een grote verzameling van gekoppelde tabellen met informatie. Over databases leer je meer in de 5e klas, wanneer je ook een programmeertaal SQL leert om bijvoorbeeld specifieke informatie uit een database te halen.

## <span style="color:brown">Webservices</span>
Veel apps en online toepassingen maken gebruik van data dat staat opgeslagen in een database. Om die data te krijgen maken ze gebruik van een <b>webservice</b>. Dat is een online dienst die beschikbaar is op een webadres. Je kunt een dataverzoek (<b>request</b>) naar de webservice sturen. De webservice haalt de data uit een database en stuurt dat vervolgens terug zodat het gebruikt kan worden in een app. Buienradar biedt als webservice live weerinformatie aan dat je in je eigen website of in een eigen app kunt gebruiken, bijvoorbeeld deze widget:

<img src="img/6.1_buienradarWidget.png">
        


Om dat mogelijk te maken wordt er gebruik gemaakt van een drie-lagen model:
- <b>Databaselaag</b>: De data in de database is opgeslagen op hardware. De database heeft software (Database Management Systeem - DBMS) dat gegevens in de database zet en eruit haalt. De DMBS zorgt er onder andere voor dat de data in de database veilig is, en dat niemand zomaar bij data kan waar ze niet bij mogen.

- <b>Servicelaag</b>: De webservice is software dat gegevens aan de DBMS vraagt en ontvangt. Deze slaat de data op in een bestand en kan deze dan doorsturen naar applicaties. Vaak worden .XML of .JSON bestandsformaten gebruikt voor data.

- <b>Applicatielaag</b>: Je browser of de app op je mobiel (beiden software) haalt de data dat nodig is uit de ontvangen bestand en toont deze op de gewenste manier op de gebruiker.

## <span style="color:brown">Stap 4. Data aan Buienradar vragen en tonen</span>
In de javascript moet je nog code schrijven om uit de JSON bestand de juiste data op te vragen.

Daarvoor moet je eerst in het JSON-bestand precies kijken hoe (met welk pad) je bij dat data komt. 


<img src="img/6.2_stap4_json.png" width="250" border="1">

Om bij temperature van meetstation Arnhem te komen moet je de volgende pad volgen:<br> 
<i>actual > stationmeasurements > 1 > temperature</i>

In javascript geef je dat pad zo aan:<br>
    `data.actual.stationmeasurements[1].temperature;`

Met de volgende code koppel je het aan je temperatuur element:
`document.getElementById("temperatuur").innerHTML = data.actual.stationmeasurements[1].temperature;`

Deze regel neem je op in je javascript-bestand <i>script.js</i>:


<img src="img/6.2_stap4_js.png">



Wil je er nog wat tekst aan de temperatuur toevoegen, dan kan dat met een `+`, bijvoorbeeld:
`document.getElementById("temperatuur").innerHTML = data.actual.stationmeasurements[1].temperature + "°C";`
